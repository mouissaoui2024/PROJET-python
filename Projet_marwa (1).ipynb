```python
data1.head()
```

```python
data2.head()
```

```python
data=pd.merge(data1, data2, on = 'Customer ID', how = 'inner')
data.head()
```

#Visualisation

```python
#importer la fonction pyplot de matplotlib pour la visualisation 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.bar(height=data['Salary'], x=data['Region'])
```

```python
#ordonner les barres de plus grand au plus petit
#augmenter la taille de la fenètre graphique  
plt.figure(figsize=(10,6))

# ordonner les barres
data_sorted = data.sort_values('Salary', ascending=False)
plt.bar(height=data_sorted['Salary'], x=data_sorted['Region'])

#Ajouter un titre et nommé les axes du graphique
plt.title("Salaire_total réalisés par Region", fontname='arial', fontsize=18)
plt.xlabel("Region", color='b', fontweight='bold',fontsize=12)
plt.ylabel("Salaire", color='red', fontweight='bold', fontsize=12)

# Ajouter un quadriage à notre graphique 
#plt.grid()

# fixerles limites des axes
plt.ylim(0,210000)
# etaler les intervalles
plt.yticks([0,30000, 60000, 90000,120000,150000,180000,210000])

#affichage textuelle des valeurs
ca=data_sorted['Salary'].tolist()
for i in range(4):
    plt.text(i-0.2, ca[i]+1000, round(ca[i],1))
plt.show()




plt.show()
```

```python
#distribution des salaries( par gendre ) par region 
sns.countplot(x="Region",
            hue="Gender",
            hue_order=["Female", "Male"],
            data=data,
            palette="tab10")
plt.xlabel('salaire')
plt.ylabel('nombre')

```

```python
#Graphique linéaire pour illustrer les tendances temporelles des salaires en fonction d'age

plt.figure(figsize=(5, 5))
sns.lineplot(x='Age', y='Salary', data=data)
plt.title('Tendance temporelle des salaires en fontion d_age')
plt.xlabel('Age')
plt.ylabel('Salaire')
plt.show()
```

```python
#Graphique linéaire pour illustrer les tendances temporelles des salaires en fonction de temps

plt.figure(figsize=(5, 5))
sns.lineplot(x='Date Joined', y='Salary', data=data)
plt.title('Tendance temporelle des salaires en fontion du temps')
plt.xlabel('Temps')
plt.ylabel('Salaire')
plt.show()
```

```python
 #Diagramme en boîte pour visualiser la distribution des salaires par genre 
plt.figure(figsize=(4, 4))
sns.boxplot(x='Gender', y='Salary', data=data)
plt.title('Distribution des salaires par genre')
plt.xlabel('Genre')
plt.ylabel('Salaire')
plt.show()
```
